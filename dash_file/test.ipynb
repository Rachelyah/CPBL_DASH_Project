```python
#iloc：先抓索引，再抓欄位

import datasource
import pandas as pd

data = datasource.lastest_datetime_data()
pd_data = pd.DataFrame(data, columns=['站點名稱', '更新時間', '行政區', '地址', '總數量', '可借', '可還'])

series1 = pd_data.iloc[0] #提供索引值抓出相關資料
series1.index   #類似list，資料型別MultiIndex，可以用for in迴圈

#嘗試用for in迴圈抓出所有索引欄位的值
for indexName in series1.index:
    print(indexName) 
```

```python
pd_data1 = pd_data.reset_index(names='序號')
pd_data1['站點名稱'] = pd_data1['站點名稱'].map(lambda name:name[11:]) #出來的格式會是series
```
